```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv('2023-08-24_actuator_6.csv')
```

```python
type(dataframe)
```

```python
dataframe['Movement']
```

```python
dataframe.head(10)
```

```python
# filter the dataframe for position
filtered_df = dataframe[dataframe['Movement'] == 'Home']
```

```python
filtered_df.head()
```

```python
plt.figure(figsize=(12, 6))  # Optional: Set the figure size

plt.plot(filtered_df['Time'], filtered_df['position'], label='Value')

plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.title('Time Series Plot')
plt.legend()

# plt.grid(True)  # Optional: Add gridlines

plt.show()
```
